```python
import pandas as pd
boston_data = pd.DataFrame()
for year in range(2001, 2013):
    print("reading year -",year)
    site = 'https://raw.githubusercontent.com/llimllib/bostonmarathon/master/results/{}/results.csv'.format(year)
    data = pd.read_csv(site, sep=',')
    data['year']=year
    print("Shape -",data.shape)
    boston_data = boston_data.append(data)

```

```python
print(boston_data.shape)
print(boston_data['year'].value_counts())
print(list(data.columns.values))
```

```python
# boston_data.to_parquet('boston.parquet')
# boston_data.to_parquet('boston.parquet')
import pyarrow as pa
import pyarrow.parquet as pq

boston_data['official'].replace('-', 0, inplace=True)
# boston_data['net'].replace('-', 0, inplace=True)
# boston_data['official'] = boston_data['official'].astype(float)
boston_data['net'] = boston_data['net'].astype(float)
# boston_data['age'] = boston_data['age'].astype(str)
# boston_data['year'] = boston_data['age'].astype(str)
boston_data.dtypes

table = pa.Table.from_pandas(boston_data)
pq.write_table(table, 'boston.parquet')
```
